```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris_new.csv")
df.head()
```

# Plotting graph For IRIS Dataset Using Seaborn Library And matplotlib.pyplot library

## A. Creating and customizing line charts using Matplotlib

```python
plt.figure(figsize=(13,6))
plt.plot(df["SepalLengthCm"], c="b", marker=".")
plt.xlabel("IDs", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.ylabel("Sepal Length in cm", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.title("Line chart for Sepal Length", 
          fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman", "style": "italic"})
plt.xticks(np.arange(0, 155, 5), rotation = 45)
plt.yticks(np.arange(4, 8.6, 0.5))
plt.grid()
plt.show()
```

## B. Visualizing relationships between two or more variables using scatter plots

```python
color_map = {
    "Iris-setosa": "#d4111b",
    "Iris-virginica": "#0313fc",
    "Iris-versicolor": "#48f514"
}

df["color_by_species"] = df["Species"].map(color_map)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], c=df["color_by_species"])
plt.xlabel("Sepal Length", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.ylabel("Sepal Width", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.title("Scatter chart for Sepal Length & Sepal Width", 
          fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman", "style": "italic"})
plt.show()
```

## C. Studying distributions of variables using histograms & bar charts

```python
# Histogram
plt.figure(figsize=(8,6))
plt.hist(df["SepalLengthCm"], bins=np.arange(4,8.5,0.4), histtype="bar")
plt.xticks(np.arange(4,8.5,0.4))
plt.xlabel("Sepal Length", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.ylabel("Number of frequency", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.title("Histogram for Sepal Length", 
          fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman", "style": "italic"})
plt.show()
```

```python
# bar chart
mean = df.groupby("Species")["SepalLengthCm"].mean()
plt.figure(figsize=(8,10))
plt.bar(mean.index, mean.values, color = ["#fc8505", "#fcf005", "#2afc05"], label = mean.index, width=0.4, edgecolor="red")
plt.yticks(np.arange(0,8,0.5))
plt.xlabel("Species", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.ylabel("Mean of Sepal Length", 
           fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})
plt.title("Bar chart for Sepal Length Average", 
          fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman", "style": "italic"})
plt.legend(title="Species")
plt.show()
```

## D. Visualizing two-dimensional data using heatmaps

```python
corr_df = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
corr_df.corr()
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(corr_df.corr(), vmin = -0.6, vmax = 1, cmap="Blues", annot=True)
plt.title("Heat map of Species data", 
          fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman", "style": "italic"})
plt.show()
```

## F. Plotting multiple Matplotlib and Seaborn charts in a grid

```python
fig, axes = plt.subplots(2,2, figsize=(14,10))

axes[0,0].plot(df["SepalLengthCm"], c="blue")
axes[0,0].set_title("Line chart of Sepal Length",
                    fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})

axes[0,1].bar(df.groupby("Species")["SepalLengthCm"].max().index, height=df.groupby("Species")["SepalLengthCm"].max().values,
              color=['#f0e02e', '#33FF57', '#3357FF'],  # Example custom colors
              width=0.5, edgecolor = "red")
axes[0,1].set_title("bar chart of species",
                   fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})

axes[1,0].hist(df["SepalLengthCm"], color = "#0ff5e9", edgecolor = "red")
axes[1,0].set_title("Histogram of Setal Length",
                   fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})

axes[1,1].scatter(df["PetalLengthCm"], df["PetalWidthCm"], color = "#7c11ba")
axes[1,1].set_title("Scatter plot of Petal Length and Width",
                   fontdict={"fontsize": 12, "fontweight": "bold", "fontname": "Times New Roman"})

plt.suptitle('Iris dataset charts', fontsize=20, fontweight='bold', fontname='Times New Roman')

plt.show()
```
